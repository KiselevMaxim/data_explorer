There are no Avengers befor 1960s.   
Need to remove issue records.
The cleared dataset is **true_avengers**

```python
true_avengers = avengers[avengers["Year"] >= 1960]
```

Let's count deaths

```python
def clean_deaths(row):
    num_deaths = 0
    columns = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
    
    for c in columns:
        death = row[c]
        if pd.isnull(death) or death == 'NO':
            continue
        elif death == 'YES':
            num_deaths += 1
    return num_deaths

true_avengers1=true_avengers.copy()
true_avengers1['Deaths'] = true_avengers1.apply(clean_deaths, axis=1)
true_avengers = true_avengers1
true_avengers
```
